# making dictionary to pass in table name fact_atliqo_metrics

```python

a = dict(zip(dim_cities.city_code,dim_cities.city_name))
b = dict(zip(dim_plan.plan,dim_plan.plan_description))
c = dict(zip(fact_market_share.city_code,fact_market_share.tmv_city_crores))
d = dict(zip(fact_market_share.city_code,fact_market_share.company))
e = dict(zip(fact_market_share.city_code,fact_market_share.ms_pct))
```

```python
fact_atliqo_metrics['city_name'] = fact_atliqo_metrics['city_code'].map(a)
fact_atliqo_metrics['tmv_city_crores'] = fact_atliqo_metrics['city_code'].map(c)
fact_atliqo_metrics['company'] = fact_atliqo_metrics['city_code'].map(d)
fact_atliqo_metrics['ms_pct'] = fact_atliqo_metrics['city_code'].map(e)
```

```python
fact_atliqo_metrics
```

# making dict again to pass entries in fact_plan_revenues

```python
fact_plan_revenue['plan_description'] = fact_plan_revenue['plans'].map(b)
```

```python
fact_plan_revenue
```

```python
dim_date
```

```python
f = dict(zip(dim_date.date,dim_date.month_name))
g = dict(zip(dim_date.date,dim_date['before/after_5g']))
h = dict(zip(dim_date.date,dim_date.time_period))
```

```python
fact_atliqo_metrics['month_name'] = fact_atliqo_metrics['date'].map(f)
fact_atliqo_metrics['before/after_5g'] = fact_atliqo_metrics['date'].map(g)
fact_atliqo_metrics['time_period'] = fact_atliqo_metrics['date'].map(h)
```

# fact_atliqo_metrics merge on basis of city_code and date with fact_plan_revenue

```python
df = pd.merge(fact_atliqo_metrics,fact_plan_revenue,how='inner',on=['city_code','date'])
```
