**making dic to mapping room category in fact_bookings**

```python
dict(zip(dim_rooms.room_id, dim_rooms.room_class ))
```

```python
fact_bookings['room_class'] = fact_bookings['room_category'].map(dict(zip(dim_rooms.room_id, dim_rooms.room_class )))
```

```python
fact_bookings.head()
```

**merging table dim hotels by making dict on property id with fact_bookings**

```python
a=dict(zip(dim_hotels['property_id'],dim_hotels['property_name']))
b=dict(zip(dim_hotels['property_id'],dim_hotels['category']))
c=dict(zip(dim_hotels['property_id'],dim_hotels['city']))
```

```python
fact_bookings['property_name'] = fact_bookings['property_id'].map(a)
fact_bookings['category'] = fact_bookings['property_id'].map(b)
fact_bookings['city'] = fact_bookings['property_id'].map(c)
```

**merging table dim date by making dict on date column with fact_bookings**

```python
e=dict(zip(dim_date['date'],dim_date['mmm yy']))
f=dict(zip(dim_date['date'],dim_date['week no']))
g=dict(zip(dim_date['date'],dim_date['day_type']))
```

```python
fact_aggregated_bookings['mmm yy'] = fact_aggregated_bookings['check_in_date'].map(e)
fact_aggregated_bookings['week no'] = fact_aggregated_bookings['check_in_date'].map(f)
fact_aggregated_bookings['day_type'] = fact_aggregated_bookings['check_in_date'].map(g)
```

```python
fact_aggregated_bookings
```

# changing date format from 1/may/22 to 1/5/22

```python
fact_aggregated_bookings['check_in_date']=pd.to_datetime(fact_aggregated_bookings['check_in_date'])
```

```python
fact_aggregated_bookings.to_csv('fact_aggregated_bookings_new_df.csv')
```

## making final table by joining and concating all tables

```python
df = pd.concat([fact_bookings,fact_aggregated_bookings],join = 'outer', axis = 1)
```
