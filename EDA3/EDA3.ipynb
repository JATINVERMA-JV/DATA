```python
dim_customers = pd.read_csv("D:/Downloads/C2 Input for participants/C2 Input for participants/dim_customers.csv")
dim_date = pd.read_csv("D:/Downloads/C2 Input for participants/C2 Input for participants/dim_date.csv")
dim_products = pd.read_csv("D:/Downloads/C2 Input for participants/C2 Input for participants/dim_products.csv")
dim_targets_orders = pd.read_csv("D:/Downloads/C2 Input for participants/C2 Input for participants/dim_targets_orders.csv")
fact_order_lines = pd.read_csv("D:/Downloads/C2 Input for participants/C2 Input for participants/fact_order_lines.csv")
fact_orders_aggregate = pd.read_csv("D:/Downloads/C2 Input for participants/C2 Input for participants/fact_orders_aggregate.csv")
```

```python
a = dict(zip(dim_customers.customer_id,dim_customers.customer_name))
b = dict(zip(dim_customers.customer_id,dim_customers.city))
```

```python
dim_targets_orders['customer_name'] = dim_targets_orders['customer_id'].map(a)
dim_targets_orders['city'] = dim_targets_orders['customer_id'].map(b)
```

```python
c = dict(zip(dim_products.product_id,dim_products.product_name))
d = dict(zip(dim_products.product_id,dim_products.category))
```

```python
fact_order_lines['product_name'] = fact_order_lines['product_id'].map(c)
fact_order_lines['category'] = fact_order_lines['product_id'].map(d)
```

```python
df =pd.merge(fact_order_lines,fact_orders_aggregate ,how='outer',on='order_id')
```
